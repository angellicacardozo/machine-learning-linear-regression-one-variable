```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('ex1data1', delimiter=',')
x = data[:,[0]]
y = data[:,[1]]
area = np.pi*3
colors = (0,0,0)
```

```python
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.xlabel('Population of city in 10,000s')
plt.ylabel('Profit in $10,000')
plt.show()
```
